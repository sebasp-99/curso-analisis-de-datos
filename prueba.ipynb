2.	Elabora un programa que, dados los catetos de un triángulo rectángulo, calcule el valor numérico de la hipotenusa:

```python
cateto_a = float(input("Ingresa la longitud del cateto a: "))
cateto_b = float(input("Ingresa la longitud del cateto b: "))
hipotenusa = (cateto_a**2 + cateto_b**2)**0.5
print("La hipotenusa es:", hipotenusa)
```
